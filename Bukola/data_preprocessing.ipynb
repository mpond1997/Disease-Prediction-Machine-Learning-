```python
import pandas as pd

# Specify the file path to your CSV file
file_path = "diabetes_012_health_indicators_BRFSS2015.csv"

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame to verify the data has been read correctly
df.head()
```

```python
# Define mapping dictionaries for each column
mapping_diabetes_012 = {0: 'no diabetes', 1: 'prediabetes or diabetes'}
mapping_high_bp = {0: 'no high BP', 1: 'high BP'}
mapping_high_chol = {0: 'no high cholesterol', 1: 'high cholesterol'}
mapping_chol_check = {0: 'no cholesterol check in 5 years', 1: 'yes cholesterol check in 5 years'}
mapping_smoker = {0: 'no', 1: 'yes'}
mapping_stroke = {0: 'no', 1: 'yes'}
mapping_heart_disease = {0: 'no', 1: 'yes'}
mapping_phys_activity = {0: 'no', 1: 'yes'}
mapping_fruits = {0: 'no', 1: 'yes'}
mapping_veggies = {0: 'no', 1: 'yes'}
mapping_hvy_alcohol_consump = {0: 'no', 1: 'yes'}
mapping_any_healthcare = {0: 'no', 1: 'yes'}
mapping_no_docbc_cost = {0: 'no', 1: 'yes'}
mapping_gen_hlth = {1: 'excellent', 2: 'very good', 3: 'good', 4: 'fair', 5: 'poor'}
mapping_diff_walk = {0: 'no', 1: 'yes'}
mapping_sex = {0: 'female', 1: 'male'}
mapping_education = {1: 'Never attended school or only kindergarten',
                    2: 'Grades 1 through 8 (Elementary)',
                    3: 'Grades 9 through 11 (Some high school)',
                    4: 'Grade 12 or GED (High school graduate)',
                    5: 'College 1 year to 3 years (Some college or technical school)',
                    6: 'College 4 years or more (College graduate)'}
mapping_age = {1: '18-24', 2: '25-29', 3: '30-34', 4: '35-39', 5: '40-44', 6: '45-49', 7: '50-54', 8: '55-59',
               9: '60-64', 10: '65-69', 11: '70-74', 12: '75-79', 13: '80 or older'}
mapping_income = {1: 'less than $10,000', 2: '$10,000 - $14,999', 3: '$15,000 - $19,999', 4: '$20,000 - $24,999',
                  5: '$25,000 - $34,999', 6: '$35,000 - $49,999', 7: '$50,000 - $74,999', 8: '$75,000 or more'}

# Apply label encoding using map() function for each column
df['Diabetes_012'] = df['Diabetes_012'].map(mapping_diabetes_012)
df['HighBP'] = df['HighBP'].map(mapping_high_bp)
df['HighChol'] = df['HighChol'].map(mapping_high_chol)
df['CholCheck'] = df['CholCheck'].map(mapping_chol_check)
df['Smoker'] = df['Smoker'].map(mapping_smoker)
df['Stroke'] = df['Stroke'].map(mapping_stroke)
df['HeartDiseaseorAttack'] = df['HeartDiseaseorAttack'].map(mapping_heart_disease)
df['PhysActivity'] = df['PhysActivity'].map(mapping_phys_activity)
df['Fruits'] = df['Fruits'].map(mapping_fruits)
df['Veggies'] = df['Veggies'].map(mapping_veggies)
df['HvyAlcoholConsump'] = df['HvyAlcoholConsump'].map(mapping_hvy_alcohol_consump)
df['AnyHealthcare'] = df['AnyHealthcare'].map(mapping_any_healthcare)
df['NoDocbcCost'] = df['NoDocbcCost'].map(mapping_no_docbc_cost)
df['GenHlth'] = df['GenHlth'].map(mapping_gen_hlth)
df['DiffWalk'] = df['DiffWalk'].map(mapping_diff_walk)
df['Sex'] = df['Sex'].map(mapping_sex)
df['Education'] = df['Education'].map(mapping_education)
df['Age'] = df['Age'].map(mapping_age)
df['Income'] = df['Income'].map(mapping_income)

# Display the transformed DataFrame
df.head()
```

```python
# Specify the file path where you want to save the CSV file
output_file_path = "transformed_diabetes_data.csv"

# Export the DataFrame to a new CSV file
df.to_csv(output_file_path, index=False)

# Print a message to confirm that the file has been saved
print("Data has been successfully exported and saved as:", output_file_path)
```
