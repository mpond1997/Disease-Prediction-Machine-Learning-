```python
import pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
import tensorflow as tf
from sklearn.preprocessing import StandardScaler
```

```python
diabetes_df = pd.read_csv(
    Path("diabetes_012_health_indicators_BRFSS2015 (1).csv")
)
diabetes_df.head(20)
```

```python
diabetes_df = diabetes_df.drop(columns= ['Education', 'Income'])
diabetes_df.head(20)
```

```python
diabetes_df.nunique()
```

```python
BMI_unique_values = diabetes_df['BMI'].value_counts()
print(BMI_unique_values)
```

```python
BMI_larger_values = BMI_unique_values[BMI_unique_values > 500]
print(BMI_larger_values)
```

```python
cutoff_value = 1000

BMI_types_to_replace = BMI_unique_values[BMI_unique_values <= cutoff_value].index.tolist()

for BMI in BMI_types_to_replace:
    diabetes_df['BMI'] = diabetes_df['BMI'].replace(BMI,"Other")

diabetes_df['BMI'].value_counts()
```

```python
MentHlth_unique_values = diabetes_df['MentHlth'].value_counts()
print(MentHlth_unique_values)
```

```python
cutoff_value = 1000

MentHlth_types_to_replace = MentHlth_unique_values[MentHlth_unique_values <= cutoff_value].index.tolist()

for Mh in MentHlth_types_to_replace:
    diabetes_df['MentHlth'] = diabetes_df['MentHlth'].replace(Mh,"Other")

diabetes_df['MentHlth'].value_counts()
```

```python
PhysHlth_unique_values = diabetes_df['PhysHlth'].value_counts()
print(PhysHlth_unique_values)
```

```python
cutoff_value = 1000

PhysHlth_types_to_replace = PhysHlth_unique_values[PhysHlth_unique_values <= cutoff_value].index.tolist()

for Ph in PhysHlth_types_to_replace:
    diabetes_df['PhysHlth'] = diabetes_df['PhysHlth'].replace(Ph,"Other")

diabetes_df['PhysHlth'].value_counts()
```

```python
diabetes_df = pd.get_dummies(diabetes_df, columns=['MentHlth'])
```

```python
X = diabetes_df.drop(columns=['Diabetes_012'])
y = diabetes_df['Diabetes_012']
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
nm_model = tf.keras.models.Sequential()

# First hidden layer
nm_model.add(tf.keras.layers.Dense(units=8, activation="relu", input_dim=73))

# Second hidden layer
nm_model.add(tf.keras.layers.Dense(units=6, activation="relu"))

# Output layer
nm_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nm_model.summary()
```

```python
# Compile the model
nm_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nm_model.fit(X_train_scaled, y_train, epochs=100)
```

```python
model_loss, model_accuracy = nm_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
